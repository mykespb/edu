Цифры в тексте
==============================

**Какая цифра чаще всего встречается в заданном тексте?**

Версия 1.2 от 2022-01-14

Уточнения:

1. текст может быть задан в тексте программы (в начале), введён с клавиатуры, из файла и т.п.;
2. текст может быть пустым или нет...
3. но он разумного размера;
4. ответ может содержать только информацию о том, какая цифра самая частая и сколько раз она встречается, а может быть и более  развёрнутым,
5. но он должен быть кратким, понятным, конкретным;
6. цифры - только десятичные;
7. если другие знаки есть в тексте, они не учитываются;
8. можно выполнять расчёты по одной строке за раз или сразу по нескольким.
9. если несколько цифр оказываются самыми частыми, то можно рассказать об одной любой, а можно и про все - как угодно.

После изучения каждой новой темы хорошо бы заново решить эту задачу с использованием новых изученных возможностей.


```python
# примеры текстов:

text1 = "hello 123 this text 989 777 has digits"
text2 = "каких 11 цифр 222 тут 3333 больше 444444 всего?"
text3 = "найдите здесь цифры!"
text4 = ""
```

```python
# версия 1 -- только встроенные возможности питона, без подключения модулей

текст = text1

цифры = "0123456789"

самая_цифра = ''
самых_цифр  = 0

for цифра in цифры:    
    этих_цифр = 0
    for буква in текст:
        if буква == цифра:
            этих_цифр += 1
            if этих_цифр > самых_цифр:
                самых_цифр = этих_цифр
                самая_цифра = цифра
            
print(f"получилось: {самая_цифра=}, {самых_цифр=}")
```

```python
# версия 2 -- только встроенные возможности питона, без подключения модулей, но с функцией

def often(текст):

    цифры = "0123456789"

    самая_цифра = ''
    самых_цифр  = 0

    for цифра in цифры:    
        этих_цифр = 0
        for буква in текст:
            if буква == цифра:
                этих_цифр += 1
                if этих_цифр > самых_цифр:
                    самых_цифр = этих_цифр
                    самая_цифра = цифра

    print(f"получилось: {самая_цифра=}, {самых_цифр=}")
```

```python
often(text1)
often(text2)
```

```python
# версия 2.1 -- только встроенные возможности питона, 
# без подключения модулей, но с функцией
# и с проверкой входных данных

def often(текст):

    print("\nисходная строка:", текст)
    
    if len(текст) == 0:
        print("входная строка пустая. разбор отменён!")
        return
    
    цифры = "0123456789"

    самая_цифра = ''
    самых_цифр  = 0

    for цифра in цифры:    
        этих_цифр = 0
        for буква in текст:
            if буква == цифра:
                этих_цифр += 1
                if этих_цифр > самых_цифр:
                    самых_цифр = этих_цифр
                    самая_цифра = цифра

    if самых_цифр:                   
        print(f"получилось: {самая_цифра=}, {самых_цифр=}")
    else:
        print("в строке вообще нет цифр")
```

```python
often(text1)
often(text2)
often(text3)
often(text4)
```
