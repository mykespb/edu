```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Генерируем случайные данные
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Настройка стилей Seaborn
sns.set_theme(style="darkgrid")

# Построение графика
plt.figure(figsize=(8, 4))
plt.plot(x, y, label="Синусоида", color="purple", linewidth=2)
plt.title("Простой график в Colab: синусоида")
plt.xlabel("Ось X")
plt.ylabel("Ось Y")
plt.legend()
plt.show()
```

```python
dot_code = """
graph CanadaBorders {
    node [shape=box, style=filled, fontname="Arial"];

    // ==========================================
    // ТЕРРИТОРИИ (Светло-серый цвет)
    // ==========================================
    node [color=lightgrey];
    Yukon [label="Юкон"];
    NWT [label="Северо-Западные территории"];
    Nunavut [label="Нунавут"];

    // ==========================================
    // ПРОВИНЦИИ (Голубой цвет)
    // ==========================================
    node [color=lightblue];
    BC [label="Британская Колумбия"];
    Alberta [label="Альберта"];
    Saskatchewan [label="Саскачеван"];
    Manitoba [label="Манитоба"];
    Ontario [label="Онтарио"];
    Quebec [label="Квебек"];
    NL [label="Ньюфаундленд и Лабрадор"];
    NB [label="Нью-Брансуик"];
    NS [label="Новая Шотландия"];
    PEI [label="Остров Принца Эдуарда"];

    // ==========================================
    // СУХОПУТНЫЕ ГРАНИЦЫ (Сплошные линии)
    // ==========================================
    Yukon -- NWT;
    Yukon -- BC;

    NWT -- Nunavut;
    NWT -- BC;
    NWT -- Alberta;
    NWT -- Saskatchewan;
    NWT -- Manitoba;

    Nunavut -- Saskatchewan; // Точка "Четыре угла"
    Nunavut -- Manitoba;
    Nunavut -- NL;          // Остров Киллиник

    BC -- Alberta;
    Alberta -- Saskatchewan;
    Saskatchewan -- Manitoba;
    Manitoba -- Ontario;
    Ontario -- Quebec;

    Quebec -- NL;
    Quebec -- NB;
    NB -- NS;

    // ==========================================
    // ИСКУССТВЕННЫЕ И МОРСКИЕ СВЯЗИ
    // ==========================================
    // Мост Конфедерации
    NB -- PEI [style=dashed, color=darkgreen, label="мост"];

    // Морские границы (Залив Святого Лаврентия и др.)
    edge [style=dashed, color=blue, label="вода"];
    Quebec -- PEI;
    Quebec -- NS;
    NL -- NS;          // Пролив Кабота
    PEI -- NS;         // Пролив Нортамберленд
    Nunavut -- Quebec; // Гудзонов пролив
}
"""

graphviz.Source(dot_code)
```
