```python
import random

# Генерируем список из 20 случайных чисел
random_list = [random.randint(1, 1000) for _ in range(20)]

print("Исходный список:", random_list)
print("\nДлина списка: ", len(random_list))
```

```python
import time

# Декоратор для измерения времени выполнения функции
def timer(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f"\nВремя выполнения функции '{func.__name__}': {end_time - start_time:.6f} секунд")
        return result
    return wrapper

# Реализация сортировки пузырьком (Bubble Sort) с декоратором
@timer
def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        for j in range(n - 1 - i):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

# Создаем копию random_list для тестирования функции сортировки
list_to_sort = list(random_list)

# Выполняем сортировку, время будет измерено декоратором
sorted_list_manual = bubble_sort(list_to_sort)

print(f"\nОтсортированный список (вручную): {sorted_list_manual}")
```

```python
import random

# Определяем типы животных, их виды и диапазоны длин
animal_types = {
    'кошка': {
        'species': ['Мейн-кун', 'Сиамская', 'Персидская', 'Британская', 'Сфинкс'],
        'tail_length_range': (20, 40), # cm
        'fur_length_range': (1, 15),  # cm
        'body_length_range': (30, 100) # cm
    },
    'собака': {
        'species': ['Лабрадор', 'Пудель', 'Чихуахуа', 'Немецкая овчарка', 'Бульдог'],
        'tail_length_range': (10, 50), # cm
        'fur_length_range': (1, 20),  # cm
        'body_length_range': (20, 120) # cm
    },
    'хомяк': {
        'species': ['Джунгарский', 'Сирийский', 'Роборовского'],
        'tail_length_range': (1, 3),   # cm
        'fur_length_range': (1, 2),    # cm (rounding up from potential 0.5)
        'body_length_range': (5, 20)   # cm
    }
}

data = []
num_animals = 50

for _ in range(num_animals):
    common_name = random.choice(list(animal_types.keys()))
    details = animal_types[common_name]

    species_name = random.choice(details['species'])
    tail_length = random.randint(*details['tail_length_range'])
    fur_length = random.randint(*details['fur_length_range'])
    body_length = random.randint(*details['body_length_range'])

    data.append([common_name, species_name, tail_length, fur_length, body_length])

print(f"Создана таблица 'data' с {len(data)} записями.")
print("Первые 5 записей:")
for i in range(min(5, len(data))):
    print(data[i])

print("\nПоследние 5 записей:")
for i in range(max(0, len(data) - 5), len(data)):
    print(data[i])
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Преобразуем список списков в pandas DataFrame для удобства
df = pd.DataFrame(data, columns=['общее название', 'название вида', 'длина хвоста', 'длина шерсти', 'длина тела'])

# Определяем признаки (X) и целевую переменную (y)
# Признаки: длина хвоста, длина шерсти, длина тела
X = df[['длина хвоста', 'длина шерсти', 'длина тела']]
# Целевая переменная: общее название
y = df['общее название']

# Кодируем целевую переменную (общее название) в числовой формат
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Разделяем данные на обучающую и тестовую выборки
# test_size=0.2 означает, что 20% данных пойдут на тестирование, 80% на обучение
# random_state для воспроизводимости результатов
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

print("Данные успешно подготовлены для обучения:")
print(f"Размер обучающей выборки (признаки): {X_train.shape}")
print(f"Размер обучающей выборки (целевая переменная): {y_train.shape}")
print(f"Размер тестовой выборки (признаки): {X_test.shape}")
print(f"Размер тестовой выборки (целевая переменная): {y_test.shape}")
print("\nПример закодированных целевых переменных (0: кошка, 1: собака, 2: хомяк - порядок может варьироваться):", label_encoder.classes_)

# Обучение и оценка модели с оптимальным количеством соседей (best_n_neighbors = 2)
# Используем значение best_n_neighbors, которое мы нашли ранее
knn = KNeighborsClassifier(n_neighbors=best_n_neighbors)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"\nТочность модели KNeighborsClassifier с n_neighbors={best_n_neighbors} на тестовой выборке: {accuracy:.2f}")
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Определяем диапазон значений для n_neighbors
n_neighbors_range = range(1, 15) # От 1 до 14 соседей

accuracies = []

for n in n_neighbors_range:
    # Инициализируем модель KNeighborsClassifier с текущим n_neighbors
    knn = KNeighborsClassifier(n_neighbors=n)

    # Обучаем модель на обучающих данных
    knn.fit(X_train, y_train)

    # Делаем предсказания на тестовых данных
    y_pred = knn.predict(X_test)

    # Оцениваем точность модели
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)

    print(f"n_neighbors: {n}, Точность: {accuracy:.2f}")

# Находим лучшее значение n_neighbors и соответствующую точность
best_n_neighbors = n_neighbors_range[accuracies.index(max(accuracies))]
best_accuracy = max(accuracies)

print(f"\nОптимальное количество соседей (n_neighbors): {best_n_neighbors}")
print(f"Лучшая точность: {best_accuracy:.2f}")
```
